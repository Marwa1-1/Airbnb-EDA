```python
plt.figure(figsize=(10,5))
sns.histplot(x='price',data=df,bins=30,kde=True)
plt.title('Distribution of Airbnb Listing Prices')
plt.show()
```

```python
#The x-axis represents the price values 
#while the y-axis represents the count of listings falling within each price range.
```

```python
#The distribution appears to be right-skewed,with a long tail on the higher price side.
#This indicates that while most listings are affordable, there are also a few higher-priced listings available.
#The presence of a kernel density estimation (KDE) curve provides additional insights into the shape of the distribution.
#It indicates the estimated probability density for different price ranges.
```

```python
plt.figure(figsize=(10,5))
sns.histplot(x='has_availability',data=df,bins=30)
plt.title('Distribution of Airbnb Listing has_availability')
plt.show()
```

```python
#The graph helps identify any imbalances or biases in the availability status of the listings.
#Understanding the distribution of availability can assist users in making informed decisions when booking accommodations.
#It can also provide valuable information to Airbnb hosts for optimizing listing availability based on demand patterns.
```

```python
df.corr()
```

```python
plt.figure(figsize=(25,20))
ax=sns.heatmap(df.corr(),annot=True)
```

```python
#The intensity of the color in each cell represents the strength of the correlation.
#Darker colors indicate a stronger correlation, while lighter colors indicate a weaker correlation.
```

```python
#The annotated values within each cell provide the actual correlation coefficients,
#ranging from -1.0 to 1.0, indicating the strength and direction of the correlation.
```

```python
#Correlation heatmaps are useful for identifying potential multicollinearity issues,
#where two or more variables are highly correlated with each other.
#High correlations between variables can impact the accuracy and interpretability of statistical models.
```
